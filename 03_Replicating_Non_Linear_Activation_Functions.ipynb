# Replicating non-linear activation functions

Neural Networks rather than us telling the model what to learn, we give it the tools to discover patterns in data and pattern on its own

And these tools are linear and non-linear functions

```python
import torch
import matplotlib.pyplot as plt
```

```python
# Create a tensor
A = torch.arange(-10,10,1)
A , A.dtype
```

```python
# Visualise the data
plt.grid()
plt.plot(A)
```

## Plot the ReLU Function

```python
# Apply relu function and plot the graph
plt.grid()
plt.plot(torch.relu(A))
```

```python
# write rele function

def relu(x):
  return torch.maximum(torch.tensor(0), x)

relu(A)
```

## Plot the custom ReLU Function

```python
# Plot custom relu function

plt.grid()
plt.plot(relu(A))
```

## Plot Sigmoid function

```python
# create sigmoid function

def sigmoid(x):
  return 1 / (1 + torch.exp(-x))
```

```python
# plot the graph

plt.grid()
plt.plot(sigmoid(A))
```
